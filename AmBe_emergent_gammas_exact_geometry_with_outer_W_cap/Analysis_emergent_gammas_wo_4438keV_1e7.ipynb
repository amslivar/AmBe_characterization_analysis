```python
"""
Import packages
"""
import uproot
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np

print("Your uproot version: ", uproot.__version__)
print("Info: uproot 4+ is recommended.")

#give path to the simulation output .root file
f = uproot.open('/home/amslivar/work/AmBe_v2/data/AmBe_emer_gamma/am_outcapsule_normalized.root')
f.classnames()


```

```python
f['tcapsule;16'].show()

t = f['tcapsule;16']
rq = t.arrays(library='np')
```

```python
EmergentGammas = rq['Eemission']
PrimaryGammas = rq['E0capsule']

print(len(PrimaryGammas))
EmergentGammas = np.concatenate( list(EmergentGammas), axis=0 )
PrimaryGammas = np.concatenate( list(PrimaryGammas), axis=0 )

EmergentGammas = EmergentGammas.flatten()
PrimaryGammas = PrimaryGammas.flatten()

print(type(EmergentGammas))
print(EmergentGammas)

print(len(PrimaryGammas))
```

```python
#plot primary gammas' spectrum


plt.hist(PrimaryGammas/1000, bins=100, range=(0, 1), color='indianred')
plt.xlabel('Primary gammas energy, [MeV]')
plt.ylabel('Count')
plt.xlim(0,1)
plt.yscale('log')
#plt.savefig("Primary_gammas_AmBe_wo_4438keV.png", bbox_inches ='tight')
plt.show()
```

```python
#plot detected gammas' spectrum


plt.hist(EmergentGammas/1000, bins=100, range=(0, 1), color='teal')
plt.xlabel('Emergent gammas energy, [MeV]')
plt.ylabel('Count')
plt.xlim(0,1)
plt.ylim(1,2e8)
plt.yscale('log')
#plt.savefig("Emergent_gammas_AmBe_wo_4438keV.png", bbox_inches ='tight')
plt.show()
```

```python
plt.hist(PrimaryGammas/1000, bins=100, range=(0, 1), color='indianred', label=r'Primary gammas, $10^7$ counts')
plt.hist(EmergentGammas/1000, bins=100, range=(0, 1), color='teal', alpha=0.5, label='Emergent gammas, {} counts'.format(len(EmergentGammas)))
plt.xlabel('Energy, [MeV]')
plt.ylabel('Count')
plt.legend()
plt.xlim(0,1)
plt.yscale('log')
plt.savefig("Primary_and_emergent_gammas_AmBe_wo_4438keV.png", bbox_inches ='tight')
plt.show()
```

```python
fraction = len(EmergentGammas)/len(PrimaryGammas)

errors_emergent = np.sqrt(EmergentGammas)/len(EmergentGammas)
errors_primary = np.sqrt(PrimaryGammas)/len(PrimaryGammas)

error = ((np.sum(errors_emergent)/len(EmergentGammas))+ np.sum(errors_primary)/len(PrimaryGammas))*fraction
print(error)

```
